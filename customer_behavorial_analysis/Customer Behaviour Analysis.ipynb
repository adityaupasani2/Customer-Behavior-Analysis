# Let's Analyze and Visualize

```python
df['Gender'].value_counts().reset_index()
```

```python
#Gender Count
df['Gender'].value_counts().plot(kind='pie',autopct='%1.2f%%')
plt.show()
```

```python
results = df['Age'].value_counts().reset_index()
results = results.rename(columns={'index': 'Age', 'Age': 'Count'})
results
```

```python
#Visualize AGe COunts

df['Age'].value_counts().plot(kind='bar')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
#Lets count the city
df['City'].value_counts().reset_index()
```

```python
#Visualizing Cities

df['City'].value_counts().plot(kind='bar')
plt.title('City Count Distribution')
plt.xlabel('City')
plt.ylabel('Count')
plt.show()
```

```python
#Membership Counts
df['Membership Type'].value_counts()
```

```python
#Lets Visualize Membership
df['Membership Type'].value_counts().plot(kind='pie',autopct='%1.2f%%')
plt.title('Membership insights')
plt.ylabel(' ')
plt.show()
```

```python
#Satisfaction Level of Customer
df['Satisfaction Level'].value_counts()
```

```python
#Visualize

df['Satisfaction Level'].value_counts().plot(kind='pie',autopct='%1.2f%%')
plt.title('Customer Satisfaction')
plt.ylabel(" ")
plt.show()
```

```python
gender_age = df.groupby('Gender')['Age'].mean()
gender_age.plot(kind='bar')
plt.title("relation between Age and Gender")
plt.ylabel('Age')
plt.show()
```

```python
# Visualizing age and gender
gender_age_data = df[['Gender', 'Age']].value_counts().reset_index().rename(columns={0: 'count'})
gender_age_data = gender_age_data.sort_values(by='count', ascending=False)

plt.figure(figsize=(16, 6))
sns.barplot(data=gender_age_data, x='Age', y='count', hue='Gender')
plt.title("Age Distribution")
plt.show()
```

```python
gender_city=df.groupby('Gender')['City'].value_counts()
gender_city.plot.bar()
plt.title('relation between city and gender')
plt.ylabel('count')
plt.show()
```

```python
# visualize the gender && membership type
gender_membership = df[['Gender', 'Membership Type']].value_counts().reset_index().rename(columns={0: 'count'})
sns.barplot(data=gender_membership, x='Gender', y='count', hue='Membership Type')
plt.title('Insights between Gender and Membership type')
plt.show()
```

```python
#visualization btn gnder and items purchased
gender_items = df.groupby('Gender')['Items Purchased'].sum()
gender_items.plot.bar()
plt.title('Sum of Items Purchased')
plt.ylabel('count')
plt.show()
```

```python
# visualize the relation between gender and discount applied
gender_discount = df.groupby('Gender')['Discount Applied'].sum()
gender_discount.plot.bar()
plt.title('Discount and gender Insights')
plt.ylabel('No.of People')
plt.show()
```

```python
#visualzation of average spending by a people of same city
plt.figure(figsize=(9,6))
sns.barplot(data=df,x='City',y='Total Spend')
plt.title('Average Spending by a city')
plt.show()
```

```python
#memebership & average spend
sns.barplot(data=df,x='Membership Type',y='Total Spend')
plt.title('Membership and Average Spend')
plt.show()
```

```python
#Lets count the number of discount and visualize
df['Discount Applied'].value_counts()
```

```python
df['Discount Applied'].value_counts().plot(kind='pie',autopct='%1.2f%%')
plt.title('Discount Distribution')
plt.show()
```

```python
# making a new column for age group
age_grp=['21-30','31-43']
df['age_group'] = pd.cut(df.Age,bins=np.arange(15,46,15),labels=age_grp,right=True)
df.head()
```

```python
#Visualize Age Groups
df['age_group'].value_counts().plot.pie(autopct='%1.2f%%')
plt.show()
```

```python
#average Spending by age group
sns.barplot(data=df,x='age_group',y='Total Spend')
plt.title('Avg. spending by the  Different Age Group')
plt.ylabel('Average Spending')
plt.show()
```

```python
#average number of products purchased by age groups
sns.barplot(data=df,x='age_group',y='Items Purchased')
plt.title('Product Purchase by Age Groups')
plt.xlabel('Age Groups')
plt.show()
```

```python
# average Rating by Age Group
sns.barplot(data=df, x='age_group', y='Average Rating')
plt.show()
```

```python
# visualize the relation between the age group and discount applied
age_group_discount = df.groupby('age_group')['Discount Applied'].value_counts()
age_group_discount.plot.bar()
```

```python
plt.figure(figsize=(9,6))
sns.histplot(df['Days Since Last Purchase'])
plt.title('Distribution of days Since Last Product Purchased')
plt.ylabel('No of Customers')
plt.show()
```

```python
# find the relation between Satistaction Level and items purchased
satis_ip=df.groupby('Satisfaction Level')['Items Purchased'].mean()
satis_ip.plot.bar()
plt.ylabel('Avg.Item Purchased')
plt.show()
```

```python
# find the relation between Satistaction Level and average rating
satisfaction_rating = df.groupby('Satisfaction Level')['Average Rating'].mean()
satisfaction_rating.plot.bar()
plt.ylabel('Average Rating')
plt.show()
```

```python
# find the relation between Satistaction Level and average spend
satisfaction_spend = df.groupby('Satisfaction Level')['Total Spend'].mean()
satisfaction_spend.plot.bar()
plt.ylabel('Average Spending')
plt.show()
```
